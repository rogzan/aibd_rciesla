```python
# Wymagane biblioteki
import numpy as np
import pandas as pd
import os
import csv
```

```python
# Odczytuje scieżke do oryginalnego pliku w celu wczytania datasetu
original_data_path = os.getcwd()[:-13] + "original_data/tb.csv"
```

```python
with open(original_data_path) as csvfile:
    df = pd.read_csv(csvfile)
```

```python
df
```

Dane zawierają rożne informacje w kolumnach, a także wiele pustych miejsc. Aby je uporządkować należy pogrupować informacje z nagłówków i przestawić kolumny(unpivot).

```python
# Przestawienie kolumn
df1 = df
df1 = df1.drop('new_sp', axis=1)
df1 = pd.melt(df1, id_vars=["iso2","year"], var_name="key", value_name="cases").dropna()
df1.columns = ['country', 'year', 'key', 'cases']  # zmiana nazwy kolumny iso2
df1
```

```python
df1.key.value_counts().head(25)
```

```python
# Rozdzielenie wartości z kolumny key
new_columns = (df1.key.str.split("_",expand = True).rename(columns = {0: "new", 1: "type", 2: "sex_and_age"}))
new_columns['sex'] = new_columns['sex_and_age'].str[0]
new_columns['age'] = new_columns['sex_and_age'].str[1:].map({
    '04': '0-14',
    '514': '0-14',
    '014': '0-14',
    '1524': '15-24',
    '2534': '25-34',
    '3544': '35-44',
    '4554': '45-54',
    '5564': '55-64',
    '65': '65+',
    'u': 'u'
})
new_columns = new_columns[['type', 'sex', 'age']]
new_columns
```

```python
# Dodanie do tabeli nowych rozdzielonych kolumn zawierających typ, płeć i przedział wiekowy
df2 = pd.concat([df1.filter(items = ['country', 'year']), new_columns, df1.filter(items = ['cases'])], axis = 1)
df2
```

```python
# Sortowanie wartości
df3 = df2.sort_values(ascending=True,by=["country", "year", "sex", "age"])
df3
```

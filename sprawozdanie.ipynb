```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    y = x * x + 5
    return y

x1 = np.linspace(-1, 1)
x2 = np.linspace(-6, 6)
x3 = np.linspace(0, 5)

y1 = f(x1)
y2 = f(x2)
y3 = f(x3)
```

Wykres funkcji na przedziale (-1, 1):

```python
plt.plot(x1, y1)
plt.xlabel("oś x")
plt.ylabel("oś y")
plt.title("Funkcja f określona na przedziale <-1;1>")
plt.grid()
plt.show()
```

Wykres funkcji na przedziale (-6, 6):

```python
plt.plot(x2, y2)
plt.xlabel("oś x")
plt.ylabel("oś y")
plt.title("Funkcja f określona na przedziale <-6;6>")
plt.grid()
plt.show()
```

Wykres funkcji na przedziale (0, 5):

```python
plt.plot(x3, y3)
plt.xlabel("oś x")
plt.ylabel("oś y")
plt.title("Funkcja f określona na przedziale <0;5>")
plt.grid()
plt.show()
```

# 10. Jakie kategorie filmów zostały wypożyczone przez klientów?

```python
df = pd.read_sql("select distinct category.name from rental inner join inventory on rental.inventory_id = inventory.inventory_id\n" 
                 "inner join film on film.film_id = inventory.film_id\n" 
                 "inner join film_category on film_category.film_id = film.film_id\n" 
                 "inner join category on film_category.category_id = category.category_id",con=connection)
df
```

# 11. Znajdź wszystkie kategorie filmów wypożyczonych w Ameryce.

```python
df = pd.read_sql("select distinct category.name from rental inner join inventory on rental.inventory_id = inventory.inventory_id\n" 
                 "inner join film on film.film_id = inventory.film_id\n" 
                 "inner join film_category on film_category.film_id = film.film_id\n" 
                 "inner join category on film_category.category_id = category.category_id\n" 
                 "inner join customer on customer.customer_id = rental.customer_id\n" 
                 "inner join address on address.address_id = customer.address_id\n" 
                 "inner join city on city.city_id = address.city_id\n" 
                 "inner join country on country.country_id = city.country_id where country.country='United States'",con=connection)
df
```
